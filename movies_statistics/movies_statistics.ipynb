```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("../datasets/movies.csv")
```

```python
df["year"] = pd.to_datetime(df["year"])
df["year"] = df["year"].dt.year
df = df.iloc[:, 1:]
```

```python
genres = df.groupby("genre")[["genre"]].count()
genres.rename(columns={"genre": "quantity"}, inplace=True)
```

```python
genres.plot(kind="bar", ylabel="Quantity", title="Movies by genres")
```

```python
plt.clf()
```

```python
genres.sort_values("quantity", ascending=False, inplace=True)
top_3_genres = genres.iloc[:3]
top_3_genres = top_3_genres.index.to_list()
```

```python
plt.hist(df[df["genre"] == top_3_genres[0]]["rating"], label=top_3_genres[0], bins=10, alpha=0.5)
plt.hist(df[df["genre"] == top_3_genres[1]]["rating"], label=top_3_genres[1], bins=10, alpha=0.5)
plt.hist(df[df["genre"] == top_3_genres[2]]["rating"], label=top_3_genres[2], bins=10, alpha=0.5)
plt.legend()
plt.xlabel("IMDB Rating")
plt.ylabel("Movies quantity")
plt.title("Comparison of top 3 genres by rating")
```

```python
plt.clf()
```
